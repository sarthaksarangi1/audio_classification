```python
import os
import glob
import librosa
import numpy as np
import pandas as pd
from sklearn.utils import resample
from itertools import chain
from more_itertools import chunked
from audiomentations import TimeStretch
import soundfile as sf

input_dir = "data"  # Directory containing your audio recordings
output_dir = "2all-balanced_data"  # Directory to save the balanced dataset
os.makedirs(output_dir, exist_ok=True)

metadata_path = "data/metadata.csv"

# Load metadata.csv file
metadata = pd.read_csv(metadata_path)

# Define target number of instances per class
target_instances = 0

# Dictionary to store instances for each class
class_data = {}

# Iterate over each row in the metadata
for index, row in metadata.iterrows():
    file_path = row['path']
    class_label = row['label']

    # Load audio
    audio, sr = librosa.load(file_path, sr=None)

    # Append audio and class label to class_data dictionary
    if class_label not in class_data:
        class_data[class_label] = []
    class_data[class_label].append(audio)

transform = TimeStretch(
    min_rate=0.8,
    max_rate=1.25,
    p=1.0
)

batch_size = 4  # Define the desired batch size

# Perform class-wise sampling and data augmentation
for class_label, instances in class_data.items():
    # Determine the number of instances in the class
    num_instances = len(instances)

    # Split instances into smaller batches
    instance_batches = list(chunked(instances, batch_size))

    # Process each batch
    for batch_num, instance_batch in enumerate(instance_batches):
        # Initialize empty list to store balanced instances for this batch
        balanced_instances = []

        # Iterate over instances in the batch
        for instance in instance_batch:
            # Apply augmentation techniques
            augmented_instance = transform(instance, sample_rate=44100)
            
            # Append augmented instance to the list
            # balanced_instances.append(instance)
            balanced_instances.append(augmented_instance)

        # Save balanced instances to the output directory
        for i, audio in enumerate(balanced_instances):
            # Generate a unique filename for each instance
            filename = f"{class_label}_{batch_num}_{i}.wav"

            # Save the audio to the output directory
            output_path = os.path.join(output_dir, filename)
            sf.write(output_path, audio, sr)

# Generate metadata for the balanced data
balanced_metadata = []

for class_label in class_data:
    class_dir = os.path.join(output_dir, str(class_label))
    files = glob.glob(os.path.join(class_dir, "*.wav"))

    for file_path in files:
        file_name = os.path.basename(file_path)
        unique_file = os.path.splitext(file_name)[0]
        species = class_label  # Add the species information if available
        label = class_label
        subset = "aug"  # Add the subset information if available
        audio, sample_rate = librosa.load(file_path, sr=44100)
        num_frames = audio
        length = num_frames / sample_rate

        metadata_entry = {
            'file_name': file_name,
            'unique_file': unique_file,
            'path': file_path,
            'species': species,
            'label': label,
            'subset': subset,
            'sample_rate': sample_rate,
            'num_frames': num_frames,
            'length': length
        }

        balanced_metadata.append(metadata_entry)

# Create a DataFrame from the balanced metadata list
balanced_metadata_df = pd.DataFrame(balanced_metadata)

# Save the balanced metadata to a CSV file
balanced_metadata_df.to_csv(os.path.join(output_dir, 'metadata.csv'), index=False)
```

```python
import os
import csv

# Set the directory path containing the audio files
directory = 'balanced_data_1/train'

# Set the path for the metadata.csv file
csv_file = 'balanced_data_1/train/metadata.csv'

# Open the metadata.csv file in write mode
with open(csv_file, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['file_name', 'label', 'path'])

    # Iterate over the audio files in the directory
    for filename in os.listdir(directory):
        if filename.endswith('.wav'):  # Modify the file extension if needed
            file_path = 'train'
            label = filename.split('_')[0]  # Extract the first word as the label

            # Write the file_name, label, and path to the metadata.csv file
            writer.writerow([filename, label, file_path])
```
