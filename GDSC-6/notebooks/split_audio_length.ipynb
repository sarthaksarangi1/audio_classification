```python
import os
import librosa
import math
import soundfile as sf


# Function to split audio files
def split_audio(input_file, output_dir, duration=20):
    # Load the audio file
    audio, sr = librosa.load(input_file, sr=None)

    # Calculate the number of splits
    total_duration = librosa.get_duration(audio, sr)
    num_splits = math.ceil(total_duration / duration)
    # Get the original filename without the extension
    file_name = os.path.splitext(os.path.basename(input_file))[0]

    # Split the audio and save as separate files
    for i in range(num_splits):
        start = i * duration
        end = min((i + 1) * duration, total_duration)

        split_audio = audio[int(start * sr):int(end * sr)]
        output_file = os.path.join(output_dir, f"{file_name}_split_{i + 1}.wav")

        sf.write(output_file, split_audio, sr)

        print(f"Split {i + 1}: {start} s - {end} s saved as {output_file}")

# Path to the folder containing audio files
folder_path = "C:\\Users\\ssarangi\\Downloads\\GDSC-6\\balanced_data\\train"

# Output folder for the split audio files
output_folder = "C:\\Users\\ssarangi\\Downloads\\GDSC-6\\split\\train"

# Iterate over the audio files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".wav"):
        audio_path = os.path.join(folder_path, filename)
        split_audio(audio_path, output_folder)
```

```python
import os
import csv

# Set the directory path containing the audio files
directory = 'split/train'

# Set the path for the metadata.csv file
csv_file = 'split/train/metadata.csv'

# Open the metadata.csv file in write mode
with open(csv_file, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['file_name', 'label', 'path'])

    # Iterate over the audio files in the directory
    for filename in os.listdir(directory):
        if filename.endswith('.wav'):  # Modify the file extension if needed
            file_path = 'train'
            label = filename.split('_')[0]  # Extract the first word as the label

            # Write the file_name, label, and path to the metadata.csv file
            writer.writerow([filename, label, file_path])
```

```python
import os
from pydub import AudioSegment

# Folder containing the audio files
folder_path = "C:\\Users\\ssarangi\\Downloads\\GDSC-6\\split\\train"

# Desired duration in milliseconds
desired_duration = 20001

# Iterate over the files in the folder
for file_name in os.listdir(folder_path):
    file_path = os.path.join(folder_path, file_name)
    
    # Check if the file is a WAV file
    if file_name.lower().endswith(".wav"):
        # Load the audio file
        audio = AudioSegment.from_wav(file_path)
        
        # Calculate the duration of the audio file in milliseconds
        duration = len(audio)
        
        # Check if the duration is less than desired_duration
        if duration < desired_duration:
            # Calculate the amount of silence to add
            silence_duration = desired_duration - duration
            
            # Create a silent audio segment with the required duration
            silence = AudioSegment.silent(duration=silence_duration)
            
            # Concatenate the original audio with the silence
            padded_audio = audio + silence
            
            # Export the padded audio as a WAV file
            padded_audio.export(file_path, format="wav")
```
